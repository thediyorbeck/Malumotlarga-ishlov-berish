```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
df.price.describe()
```

```python
bins=[5000,10000,30000,50000]
```

```python
groups=['arzon',"o'rta","qimmat"]
```

```python
df['narx-oraliqlari']=pd.cut(df.price, bins, labels=groups)   #lebels - oraliq qiymatlarga nom berish uchun
df
```

```python
df['narx-oraliqlari'].value_counts()
```

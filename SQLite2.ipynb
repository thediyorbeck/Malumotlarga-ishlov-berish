Bo'sh jadval yaratish

```python
query="""
CREATE TABLE IF NOT EXISTS dunyoaholisi(
  rank INT,
  country TEXT PRIMARY KEY,
  density REAL,
  population REAL NOT NULL,
  area REAL NOT NULL
);
"""
cursor.execute(query)
db.commit()
```

```python
cursor.execute('SELECT name FROM sqlite_master WHERE type="table";')
cursor.fetchall()
```

```python
cursor.execute('INSERT INTO dunyoaholisi VALUES(1,"Macau",21946.4667, 658.3940, 30)')
db.commit()
```

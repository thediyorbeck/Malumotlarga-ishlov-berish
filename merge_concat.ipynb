```python
import pandas as pd
import numpy as np
```

```python
A = pd.DataFrame({'Transport': ['poezd', 'avtomabil', 'kema'],
                  'Tezligi':['70', '180', '45']})

A
```

```python
B = pd.DataFrame({'Transport': ['poezd', 'avtomabil', 'tramvay'],
                  'Narxi':['98000', '10000', '65000']})

B
```

```python
pd.merge(A,B)
```

```python
pd.merge(A,B,how='outer')
```

```python
pd.merge(A,B,how='left')
```

```python
pd.merge(A,B,how='right')
```

```python
A.rename(columns={'Transport':'Transport_turi'},inplace=True)
A
```

```python
pd.merge(A,B,left_on='Transport_turi',right_on='Transport').drop('Transport',axis=1)
```

```python
A.set_index('Transport_turi',inplace=True)
A
```

```python
pd.merge(B,A,right_index=True,left_on='Transport')
```

```python
B.set_index('Transport', inplace=True)
B
```

```python
A
```

```python
pd.merge(A,B,left_index=True,right_index=True)
```

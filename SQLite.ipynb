```python
db =sql.connect("chinook.db")
cursor = db.cursor()
```

```python
cursor.execute('SELECT name FROM sqlite_master WHERE type="table"')
cursor.fetchall()
```

```python
cursor.execute('SELECT COUNT (*) FROM playlists')
cursor.fetchone()
```

```python
cursor.execute("PRAGMA table_info(customers)")
cursor.fetchall()
```

```python
query="SELECT FirstName, LastName, Company, State, Country,Phone FROM customers LIMIT 5"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM customers ORDER BY FirstName"
df = pd.read_sql(query, db, index_col="CustomerId")
df
```

```python
query = "SELECT DISTINCT Country FROM customers"
cursor.execute(query)
cursor.fetchall()
```

```python
query = 'SELECT * FROM customers WHERE country="Brazil" and city="São Paulo" ORDER BY Lastname'
df = pd.read_sql(query, db, index_col="CustomerId")
df
```

```python
db=sql.connect('electricity.db')
cursor=db.cursor()
```

```python
cursor.execute('SELECT name FROM sqlite_master WHERE type="table";')
cursor.fetchall()
```

```python
pd.read_sql('SELECT * FROM elect',db,index_col='index')
```

```python
query='SELECT DISTINCT country FROM elect'
countries = pd.read_sql(query,db)
```

```python
'Uzbekistan' in countries['country'].values
```

```python
query='SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country="Uzbekistan" AND year>=2010'
dfuz=pd.read_sql(query,db)
dfuz
```

```python
query='SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country="Uzbekistan" AND year IN (2006,2009,2011,2015)'
pd.read_sql(query,db)
```

```python
query='SELECT * FROM elect WHERE country LIKE "%stan" AND year IN (2006,2009,2011,2015)'
pd.read_sql(query,db)
```
